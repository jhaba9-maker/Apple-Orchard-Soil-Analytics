```python
import matplotlib.pyplot as plt

plt.scatter(df["ph"], df["label"])
plt.xlabel("pH")
plt.ylabel("Cultivo (label)")
plt.title("Relación entre pH y tipo de cultivo")
plt.show()
```

```python
df.boxplot(column="N", by="label", figsize=(10,5))
plt.xticks(rotation=90)
plt.show()
```

```python
##¿Cuantos registros hay por cultivo?
df["label"].value_counts()
```

```python
## Distribución del pH
plt.figure(figsize=(8,5))
plt.hist(df["ph"], bins=20)
plt.title("Distribución del pH")
plt.xlabel("pH")
plt.ylabel("Frecuencia")
plt.show()
```

```python
## Distribución de nitrógeno (N)
plt.figure(figsize=(8,5))
plt.hist(df["N"], bins=20, color='gray')
plt.title("Distribución de Nitrógeno (N)")
plt.xlabel("N")
plt.ylabel("Frecuencia")
plt.show()
```

```python
## Correlacion entre pH y N
corr = df.corr(numeric_only=True)

corr
```

```python
## Caso de estudio: Manzanas
print(f"Número de cultivos: {df['label'].nunique()}")

cultivos = sorted(df['label'].unique())

for cultivo in cultivos:
    print(cultivo)
```
