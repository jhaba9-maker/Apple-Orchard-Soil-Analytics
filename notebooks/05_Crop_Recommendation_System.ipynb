# Notebook 5 - Crop Recommendation System

## Objetivo

Desarrollar un sistema de recomendación capaz de predecir el cultivo más adecuado utilizando el modelo Random Forest entrenado en el Notebook 4.

El sistema recibirá como entrada las características del suelo y devolverá:

- Cultivo recomendado.
- Probabilidad de la predicción.
- Las tres recomendaciones más probables.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("../data/processed/soil_ml_ready.csv")
df.head()
```

```python
## Separar variables

X = df.drop("label", axis=1)
y = df["label"]
```

```python
## Train / test split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
## Entrenar modelo

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
## Funcion 1

def create_input(N, P, K, temperature, humidity, ph, rainfall):

    total_npk = N + P + K

    return pd.DataFrame({
        "N": [N],
        "P": [P],
        "K": [K],
        "temperature": [temperature],
        "humidity": [humidity],
        "ph": [ph],
        "rainfall": [rainfall],
        "Total_NPK": [total_npk],
        "N_P_ratio": [N/P],
        "N_K_ratio": [N/K],
        "P_K_ratio": [P/K]
    })
```

```python
## Ejemplo de entrada

soil = create_input(
    N=90,
    P=42,
    K=43,
    temperature=22,
    humidity=82,
    ph=6.5,
    rainfall=120
)

soil
```

```python
## Función predicción basica

def predict_crop(model, input_data):
    return model.predict(input_data)[0]
```

```python
## Predicción

predict_crop(model, soil)
```

```python
## Predicción con confianza

def predict_crop_with_confidence(model, input_data):

    prediction = model.predict(input_data)[0]

    probs = model.predict_proba(input_data)[0]

    confidence = np.max(probs)

    return prediction, confidence
```

```python
## Probar confianza 

crop, conf = predict_crop_with_confidence(model, soil)

print(crop, conf)
```

```python
## Funcion 4 Top 3

def predict_top3_table(model, input_data):

    probs = model.predict_proba(input_data)[0]
    classes = model.classes_

    top_idx = np.argsort(probs)[::-1][:3]

    results = {
        "Rank": [],
        "Crop": [],
        "Probability": []
    }

    for i, idx in enumerate(top_idx, 1):
        results["Rank"].append(i)
        results["Crop"].append(classes[idx])
        results["Probability"].append(probs[idx])

    return pd.DataFrame(results)
```

```python
## Prueba top 3

predict_top3_table(model, soil)
```

```python
## Prueba varios escenarios

soil2 = create_input(60,30,40,25,95,6.8,200)
soil3 = create_input(40,20,25,35,30,6.2,40)
soil4 = create_input(120,50,60,24,75,6.5,110)

predict_top3_table(model, soil2)
predict_top3_table(model, soil3)
predict_top3_table(model, soil4)
```
